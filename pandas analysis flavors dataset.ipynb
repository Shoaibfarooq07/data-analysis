```python
import pandas as pd
```

```python
css = pd.read_csv("Flavors.csv")
css
```

```python
group_data=css.groupby("Base Flavor")
print(group_data)
```

```python
group_data.mean(numeric_only=True)
```

```python
css.groupby("Base Flavor").mean(numeric_only=True)
```

```python
css.groupby("Base Flavor").count()
```

```python
css.groupby("Base Flavor").min()
```

```python
css.groupby("Base Flavor").max()
```

```python
css.groupby('Base Flavor').sum(numeric_only=True)
```

```python
css.groupby('Base Flavor').agg({"Flavor Rating":['sum','mean','min','max'],"Texture Rating":['sum','mean','min','max']})
```
